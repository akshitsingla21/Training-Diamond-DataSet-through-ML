```python
from pycaret.utils import version
from pycaret.datasets import get_data

datasets= get_data("index")
```

```python
from pycaret.regression import *
diamondDataSet = get_data("diamond")

setup(data=diamondDataSet, target='Price')
cm=compare_models
```

```python

#using normalisation + outlier removal + feaature selection + pca + transformation

setup(data=diamondDataSet, target='Price', normalize=True, remove_outliers=True, transformation=True,
      normalize_method='zscore', outliers_threshold=0.05, transformation_method='yeo-johnson')

cm= compare_models()
```
